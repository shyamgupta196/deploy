We are using the halloween candy bars dataset for model. and we have to predict whether the candy has CHOCOLATE in it or NOT!!


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datasets/candy-data.csv')
```

```python
df.head()
```

we have many variables in the data !!! so we filter columns which are useful. 

```python
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
```

SUGARPERCENT IS ONLY NUMERIC COLUMN WHICH HAS LEAST CORRELATION WITH CHOCOLATE VARIABLE! LET'S DROP IT 

```python
## lets prepare data for training !!
from sklearn.model_selection import train_test_split


x = df.drop(['chocolate','competitorname','sugarpercent'],axis=1)
y = df['chocolate']

x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
## import logistic regression from scikit-learn !!
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## initialise the model and fit/train it on data
LogReg = LogisticRegression()
LogReg.fit(x_train,y_train)
```

```python
pred = LogReg.predict(x_test)
print(accuracy_score(y_test,pred))
```

We achieved a 90% accuracy just with a few lines of code !!
